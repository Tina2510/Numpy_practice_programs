10 que [real and imaginary part of array]


```python
complex_array = np.array([1+0j, 0.707106+0707106j])
print(complex_array)
real_num = complex_array.real
print(real_num)
imaginary_num = complex_array.imag
print(imaginary_num)
```

12 que[comman values]


```python
array12a = np.array([0, 10, 20, 40, 60])
array12b = np.array([10, 30, 40])
print(array12a)
print(array12b)
common_val = np.intersect1d(array12a, array12b)
print(common_val)
```

13 que[set difference]

```python
array13a = np.array([0, 10, 20, 40, 60, 80])
array13b = np.array([10, 30, 40, 50, 70, 90])
print(array13a)
print(array13b)
difference = np.setdiff1d(array13a, array13b)
print(difference)
```

14th que[exclusive -or]

```python
array14a = np.array([0, 10, 20, 40, 60, 80])
array14b = np.array([10, 30, 40, 50, 70])
print(array14a)
print(array14b)
exclusive_or = np.setxor1d(array14a, array14b)
print(exclusive_or)
```

5th que [create 2D array with 1 on border and zeros inside]

```python
import numpy as np
array5 = np.ones((5,5))
print(array5)
array5[1:-1,1:-1] = 0
print(array5)
```

6th que [program to add a border filled with zero's around existing array]

```python
old_array = np.ones((3,3))
print(old_array)
new_arr = np.zeros((5,5))
new_arr[1:-1, 1:-1] = old_array
print(new_arr)
```

7th que[ checkerboard pattern]

```python
array7 = np.zeros((8,8))
print(array7)
array7[::2,1::2] = 1
array7[1::2, ::2] = 1
print()
print(array7)
```
